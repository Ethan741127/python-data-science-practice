# 二、製作累積確診數時序資料

製作空的 DataFrame

```python
df_global_timeseries_confirmed = pd.DataFrame({'date':[],'confirmed':[]})
df_global_timeseries_confirmed
```

從開始記錄至今的日期

```python
df_confirmed.columns[4:]
```

當日累積總確診數

```python
df_confirmed['8/8/20'].sum()
```

開始製作時序資料

```python
for date in df_confirmed.columns[4:]:
    df_temp = pd.DataFrame({'date':[date], 'confirmed':[df_confirmed[date].sum()]})
    df_global_timeseries_confirmed = df_global_timeseries_confirmed.append(df_temp).reset_index(drop=True)
```

```python
df_global_timeseries_confirmed
```

```python
df_global_timeseries_confirmed.plot(x='date',y='confirmed', figsize=(8,5))
```

# 三、使用非線性迴歸預測未來確診數

```python
import numpy as np
X = df_global_timeseries_confirmed.index.to_numpy()
y = df_global_timeseries_confirmed['confirmed']
```

製作二次項特徵

```python
from sklearn.preprocessing import PolynomialFeatures
quadratic = PolynomialFeatures(degree=2)
X_poly = quadratic.fit_transform(X.reshape(-1, 1))
```

建模

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X_poly, y)
```

```python
regr.score(X_poly, y)
```

### 確診數預測圖
- 藍色實線：真實確診數
- 橘色虛線：預測未來確診數趨勢

預測未來連續50天內的確診數

```python
X_fit = np.arange(200,300,1)
```

```python
import matplotlib.pyplot as plt

plt.plot(X, y, '-',linewidth=1)
plt.plot(X_fit, regr.predict(quadratic.fit_transform(X_fit.reshape(-1, 1))),'--',  color='orange', linewidth=1)
plt.show()
```
