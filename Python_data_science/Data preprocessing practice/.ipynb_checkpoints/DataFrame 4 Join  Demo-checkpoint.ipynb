```python
import pandas as pd
df1=pd.DataFrame({'name':['ruby','natasha','apple','eva','ruth'],'price':[80,90,95,80,83]})
df1
```

```python
df2=pd.DataFrame({'name':['ruby','bill','apple','vivian','ruth'],'high':[170,166,160,165,155]})
df2
```

```python
df3 =pd.merge(df1,df2,left_on='name',right_on='name',how='inner')
df3
```

```python
df4= pd.merge(df1,df2,left_on='name',right_on='name',how='outer')
df4
```

```python
df5= pd.merge(df1,df2,left_on='name',right_on='name',how='left')
df5
```

```python
df6= pd.merge(df1,df2,left_on='name',right_on='name',how='right')
df6
```
