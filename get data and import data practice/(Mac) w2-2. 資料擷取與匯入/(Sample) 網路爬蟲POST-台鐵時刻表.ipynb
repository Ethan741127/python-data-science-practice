```python
import requests
payload = {
    "FromCity":"0",
    "FromStation":"1008",
    "FromStationName":"0",
    "ToCity":"10",
    "ToStation":"1238",
    "ToStationName":"0",
    "TrainClass":"2",
    "searchdate":"2017-06-19",
    "FromTimeSelect":"0000",
    "ToTimeSelect":"2359",
    "Timetype":"1",
}
res_post = requests.post("http://twtraffic.tra.gov.tw/twrail/TW_SearchResult.aspx",data = payload) 
print (res_post.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res_post.text,'html.parser')
print (soup.title.string)
```

```python
for train in soup.select("script"):
    print(train)
```

```python
for train in soup.select("script"):
    if('var JSONData' in train.text):
        print(train.text)
        trains = train.text
```
